```python
import pandas as pd
score_df = pd.DataFrame([[1,50,80,70,'boy',1], 
              [2,60,45,50,'boy',2],
              [3,98,43,55,'boy',1],
              [4,70,69,89,'boy',2],
              [5,56,79,60,'girl',1],
              [6,60,68,55,'girl',2],
              [7,45,70,77,'girl',1],
              [8,55,77,76,'girl',2],
              [9,25,57,60,'girl',1],
              [10,88,40,43,'girl',3],
              [11,25,60,45,'boy',3],
              [12,80,60,23,'boy',3],
              [13,20,90,66,'girl',3],
              [14,50,50,50,'girl',3],
              [15,89,67,77,'girl',3]],columns=['student_id','math_score','english_score','chinese_score','sex','class'])
score_df
```

```python
#1.找出全年級各科成績最高分與最低分
print('數學最高分',max(score_df['math_score']))
print('英文最高分',max(score_df['english_score']))
print('中文最高分',max(score_df['chinese_score']))
```

```python
#2.找出數學班平均最高的班級?
score_df.to_csv('score_df.csv')
score_math = pd.read_csv('score_df.csv', usecols =['math_score','class'])
score_math.groupby('class').mean()
```

```python
#3.分析全校女生與男生國文平均差幾分?
score_chinese = pd.read_csv('score_df.csv', usecols =['chinese_score','sex'])
score_chinese.groupby('sex').mean()
```
